```python
#아파트 매매가 상승률 그래프==============================================================================================================================
@app.callback(
    Output('f_graph','figure'),
    Output('f_country_graph','figure'),
    Input('country_drop','value'),
    Input('year_slider','value')
)
def update_graph(country,year):
    apart = apart_f[country]
    apart_country = apart_f.iloc[-1,7:]
    
    if year[1] != 2020:
        apart = apart.iloc[list(apart.index.str.contains(str(year[0]))).index(True):list(apart.index.str.contains(str((year[1]+1)))).index(True)]
    else :
        apart = apart.iloc[list(apart.index.str.contains(str(year[0]))).index(True):]
        
        
    fig = px.line(apart)
    fig.update_xaxes(ticktext=["15년1월","15년12월","16년12월","17년12월","18년12월","19년12월","20년12월"],
                     tickvals=['2015년 01월05일','2015년 12월28일','2016년 12월26일','2017년 12월25일','2018년 12월31일','2019년 12월30일','2020년 12월28일'],
                     tickangle=0)
    fig.update_layout(title = '아파트 매매가 상승률',
                      xaxis_title = '년도' , 
                      yaxis_title = '상승률',
                      title_x=0.5,
                      hovermode="x unified")
    fig.update_traces(mode="lines", hovertemplate=None)
    
    fig2 = px.bar(apart_country,color=apart_country.index)
    fig2.update_layout(title = '아파트 년도별 지역  상승률',
                      xaxis_title = '지역' , 
                      yaxis_title = '상승률',
                      title_x=0.5,
                      hovermode="x unified")
    
    

    return fig,fig2
#아파트 전세 상승률 그래프==============================================================================================================================
@app.callback(
    Output('yf_graph','figure'),
    Input('country_drop','value'),
    Input('year_slider','value')
)
def yeart_update(country,year):
    apart_year = apart_yf[country]
    if year[1] != 2020:
        apart_year = apart_year.iloc[list(apart_year.index.str.contains(str(year[0]))).index(True):list(apart_year.index.str.contains(str((year[1]+1)))).index(True)]
    else :
        apart_year = apart_year.iloc[list(apart_year.index.str.contains(str(year[0]))).index(True):]
    fig2 = px.line(apart_year)
    fig2.update_xaxes(ticktext=["15년1월","15년12월","16년12월","17년12월","18년12월","19년12월","20년12월"],
                     tickvals=['2015년 01월','2015년 12월','2016년 12월','2017년 12월','2018년 12월','2019년 12월','2020년 12월'],
                     tickangle=0)
    fig2.update_layout(title = '아파트 전세 상승률',
                      xaxis_title = '년도' , 
                      yaxis_title = '상승률',
                      title_x=0.5,
                      hovermode="x unified")
    fig2.update_traces(mode="lines", hovertemplate=None)
    return fig2
#아파트 월세 상승률 그래프==============================================================================================================================
@app.callback(
    Output('mf_graph','figure'),
    Input('country_drop','value'),
    Input('year_slider','value')
)
def month_update(country,year):
    apart_month = apart_mf[country]
    if year[1] != 2020:
        apart_month = apart_month.iloc[list(apart_month.index.str.contains(str(year[0]))).index(True):list(apart_month.index.str.contains(str((year[1]+1)))).index(True)]
    else :
        apart_month = apart_month.iloc[list(apart_month.index.str.contains(str(year[0]))).index(True):]
    fig3 = px.line(apart_month)
    fig3.update_xaxes(ticktext=["15년7월","15년12월","16년12월","17년12월","18년12월","19년12월","20년12월"],
                     tickvals=['2015년 07월','2015년 12월','2016년 12월','2017년 12월','2018년 12월','2019년 12월','2020년 12월'],
                     tickangle=0)
    fig3.update_layout(title = '아파트 월세 상승률',
                      xaxis_title = '년도' , 
                      yaxis_title = '상승률',
                      title_x=0.5,
                      hovermode="x unified")
    fig3.update_traces(mode="lines", hovertemplate=None)
    return fig3

#아파트 지수 상승률 그래프==============================================================================================================================
@app.callback(
    Output('i_graph','figure'),
    Input('country_drop','value'),
    Input('year_slider','value')
)
def update_i_graph(country,year):
    apart = apart_i[country]
    if year[1] != 2020:
        apart = apart.iloc[list(apart.index.str.contains(str(year[0]))).index(True):list(apart.index.str.contains(str((year[1]+1)))).index(True)]
    else :
        apart = apart.iloc[list(apart.index.str.contains(str(year[0]))).index(True):] 
    fig = px.line(apart)
    fig.update_xaxes(ticktext=["15년1월","15년12월","16년12월","17년12월","18년12월","19년12월","20년12월"],
                     tickvals=['2015년 01월05일','2015년 12월28일','2016년 12월26일','2017년 12월25일','2018년 12월31일','2019년 12월30일','2020년 12월28일'],
                     tickangle=0)
    fig.update_layout(title = '아파트 지수 상승률',
                      xaxis_title = '년도' , 
                      yaxis_title = '상승률',title_x=0.5)
    fig.update_traces(mode="lines", hovertemplate=None)
    return fig
    
#탭 선택 함수==============================================================================================================================    
@app.callback(
    Output('sub', 'children'),
    Input('Tabs', 'value'),
)
def render_content(tab):
    if tab == 'tab1':
        return Sub_tabs
    elif tab == 'tab2':
        return Sub_tabs2
    elif tab == 'tab3':
        return Sub_tabs
    
#서브탭 선택 함수==============================================================================================================================    
@app.callback(
    Output('tab_contents', 'children'),
    Input('Sub_tabs', 'value')
)
def render_content(tab):
    if tab == 'sub_tab1':
        return Tab1_div
    elif tab == 'sub_tab2':
        return Tab2_div,
    elif tab == 'sub_tab3':
        return Tab3_div,
    elif tab == 'sub_tab4':
        return Tab4_div
    elif tab == 'sub_tab5':
        return Tab5_div

#============================================================================================================================== 
@app.callback(
    
    Output('link1','active'),
    Output('link2','active'),
    Output('link3','active'),
    
    Output('alink1','n_clicks'),
    Output('alink2','n_clicks'),
    Output('alink3','n_clicks'),
    
    Output('alink1','style'),
    Output('alink2','style'),
    Output('alink3','style'),
    
    Input('alink1','n_clicks'),
    Input('alink2','n_clicks'),
    Input('alink3','n_clicks'),
)
def link_update(c1,c2,c3):
    if c1==1:
        return True,False,False,0,0,0,{'color':'white'},{'color':'#007bff'},{'color':'#007bff'}
    elif c2==1:
        return False,True,False,0,0,0,{'color':'#007bff'},{'color':'white'},{'color':'#007bff'}
    elif c3==1:
        return False,False,True,0,0,0,{'color':'#007bff'},{'color':'#007bff'},{'color':'white'}



app.run_server()
```
